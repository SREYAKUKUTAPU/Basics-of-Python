#Modes of opening a file
'r'-Read only: The file pointer is placed at the beginning of the file. This is the default mode.

'w'	Write only: Overwrites the file if it exists. Creates a new file if it doesn't exist.

'a'	Append only: The file pointer is placed at the end of the file. Creates a new file if it doesn't exist.

'wt'	Write only in text mode: This is the same as w since text mode is the default.

'wb'	Write only in binary mode.

'rb'	Read only in binary mode.

'rt'	Read only in text mode: This is the same as r since text mode is the default.

#with open()
The with open() statement automatically handles opening and closing the file, even if errors occur. This ensures that your file is always closed properly, which is important for preventing data loss.

```python
#by default opens in read mode
with open("/content/drive/MyDrive/Colab Notebooks/file_operations.txt") as f:
  print(f.read())
```

```python
#to check if file is closed or not use closed

f.closed
```

```python
with open("/content/drive/MyDrive/Colab Notebooks/file_operations.txt","w") as f:
  f.write("with open() in write mode")
print(f.closed) #True

with open("/content/drive/MyDrive/Colab Notebooks/file_operations.txt") as f:
  print(f.read())
  print(f.closed) #False

print(f.closed) #True
```

#Reading a file
1. read():reads the whole data
2. read(l):reads data of length l
3. tell(): tells you about the position of file handle
4. seek(): It helps to re-position your file handle
5. readiness(): reads data line by line

```python
with open("/content/drive/MyDrive/Colab Notebooks/creating_file.txt") as f:
  print(f.read(4)) #returns first 4 characters
  print(f.read(10)) # returns next 10 characters

  print(f.tell()) #tells the position of file handle
  print(f.seek(0)) #re-positions the file handle to 0
  print(f.tell())
  print(f.read(5))
  print(f.seek(0))
  for line in f:
    print(line)

  print(f.seek(0))
  lines=f.readlines() #to read the file line by line
  print(lines)
```
